假設函數f(x) = x**2 請用梯度下降法取得最小值

定義 objective function (目標函數) 用來判斷模型輸出結果是否準確的指標函數 以machine learning 的例子來說 loss function 找一權重可以使loss function 越小越好


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



# define objective function

def obj_func(x):

    return x **2


# one order difference function

def diff_func(x):

    return 2*x

# defeine gradient descend parameter

# x_start : x initial values

# epochs : cycle times

# lr : learning rate

def GD(x_start, df, epochs, lr):

    xs = np.zeros(epochs+1)

    xs[0] = x_start

    for i in range(epochs):

        x = x_start

        x += - diff_func(x)*lr

        xs[i+1] = x

        return xs
    

#hyperparameter

x_start = -5

epochs = 15

lr = 0.3

#gradient descend method

w = GD(x_start, diff_func, epochs, lr)

print(np.around(w, 2))

t = np.arange(-5, 5, 0.01)

plt.plot(t, obj_func(t), c = "b")

plt.plot(w, obj_func(w), c = "r", marker = "o", markersize = 5)

# plt.rcParams["font.sans-serif"] = ["Microsoft JhengHei"]

# plt.rcParams["axes.unicode_minus"] = False

plt.title("gradient descend method")

plt.xlabel("X", fontsize = 20)

plt.ylabel("loss function", fontsize = 20)

plt.show()



    
```
